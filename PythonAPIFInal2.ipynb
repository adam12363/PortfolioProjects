```python
#Coin trends over time 
df3 = df.groupby ('name', sort= False) [['quote.USD.percent_change_1h', 'quote.USD.percent_change_7d', 'quote.USD.percent_change_30d','quote.USD.percent_change_60d','quote.USD.percent_change_90d']].mean()
df3
```

```python
#Stacking of the data 
df4 = df3.stack()
df4
```

```python
type(df4)
```

```python
df5 = df4.to_frame(name='values')
df5
```

```python
df5.count()
```

```python
#Setting of an index due to the way the data was structured above.
index = pd.Index(range(100))

# Set the above DataFrame index object as the index
# using set_index() function
df6 = df5.reset_index()
df6
```

```python
#Changing of column name to percent change
df7 = df6.rename(columns={'level_1': 'Percent_change'})
df7
```

```python
#Changing the names of data points to make it more readable in visualisations
df8['Percent_change']= df7['Percent_change'].replace(['quote.USD.percent_change_1h','quote.USD.percent_change_7d', 'quote.USD.percent_change_30d', 'quote.USD.percent_change_60d', 'quote.USD.percent_change_90d'],['1h','7d', '30d', '60d', '90d'])
df8
```

```python
# Importing Seaborn and Matplotlib for visulisations
import seaborn as sns 
import matplotlib.pyplot as plt
plt.show()
sns.catplot(x='Percent_change', y='values', hue='name', data= df8, kind='point')
```

```python
#Creation of a dataframe to track the price of bitcoin 
df10 = df7[['name', 'Percent_change']]
df10 = df7.query("name == 'Bitcoin'")
df10['Percent_change']= df7['Percent_change'].replace(['quote.USD.percent_change_1h','quote.USD.percent_change_7d', 'quote.USD.percent_change_30d', 'quote.USD.percent_change_60d', 'quote.USD.percent_change_90d'],['1h','7d', '30d', '60d', '90d'])
df10
```

```python
#Creating a line graph to track the percent change of Bitcoin over 90 days 
sns.set_theme(style="darkgrid")
sns.lineplot(x='Percent_change', y='values', data = df10)
plt.show()
```
